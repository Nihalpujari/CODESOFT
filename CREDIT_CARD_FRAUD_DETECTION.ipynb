```python
  import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
train_data = pd.read_csv('fraudTrain.csv')
test_data = pd.read_csv('fraudTest.csv')
```

```python
train_data.head()
```

```python
train_data['is_fraud'].value_counts()
```

```python
cols_to_drop = ['Unnamed: 0','cc_num','merchant','first','last','trans_num','unix_time','street','category']
train_data.drop(columns=cols_to_drop,inplace = True)
test_data.drop(columns=cols_to_drop,inplace = True)
```

```python
train_data.columns = train_data.columns.str.strip().str.lower()
test_data.columns = test_data.columns.str.strip().str.lower()

cols_to_drop = ['unnamed: 0', 'cc_num', 'merchant', 'first', 'last', 'trans_num', 'unix_time', 'street', 'category']

print("Train Data Columns After Cleaning:")
print(train_data.columns.tolist())

print("\nTest Data Columns After Cleaning:")
print(test_data.columns.tolist())
```

```python
train_data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
le.fit(train_data['gender'])
```

```python
train_data[(train_data['gender'] == "M")]["gender"] = 0
```

```python
train_data.gender =[ 1 if value == "M" else 0 for value in train_data.gender]
test_data.gender =[ 1 if value == "M" else 0 for value in test_data.gender]
```

```python
train_data
```

```python
X_train = train_data.drop('is_fraud',axis=1)
X_test = test_data.drop('is_fraud',axis=1)
y_train = train_data['is_fraud']
y_test = test_data['is_fraud']
```

```python
from sklearn.preprocessing import StandardScaler
X_fraudTrain = X_train.select_dtypes(include=['int64', 'float64'])
X_fraudTest = X_test.select_dtypes(include=['int64', 'float64'])
scaler = StandardScaler()
X_Train_scaled = scaler.fit_transform(X_fraudTrain)
X_Test_scaled = scaler.transform(X_fraudTest)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 45)
dtc.fit(X_Train_scaled, y_train)
y_pred = dtc.predict(X_Test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
